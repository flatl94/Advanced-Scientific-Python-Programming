This is a test for using python with gpu acceleration

```python
import cupy as cp
import numpy as np
import time
def cupy_fourier_transform(array):
  start_time = time.time()
  res = cp.fft.fft2(array)
  end_time = time.time()
  calc_time = end_time - start_time
  return calc_time

def numpy_fourier_transform(array):
  start_time = time.time()
  res = np.fft.fft2(array)
  end_time = time.time()
  calc_time = end_time - start_time
  return calc_time
cp_128 = []
np_128 = []
cp_256 = []
np_256 = []
cp_512 = []
np_512 = []
cp_1024 = []
np_1024 = []
cp_2048 = []
np_2048 = []
for i in range(10):
  N = 128
  cp128 = cp.random.random(size = N*N).reshape(N,N)
  np128 = np.random.random(size = N*N).reshape(N,N)
  cp_128.append(cupy_fourier_transform(cp128))
  np_128.append(numpy_fourier_transform(np128))


  N = 256
  cp256 = cp.random.random(size = N*N).reshape(N,N)
  np256 = np.random.random(size = N*N).reshape(N,N)
  cp_256.append(cupy_fourier_transform(cp256))
  np_256.append(numpy_fourier_transform(np256))


  N = 512
  cp512 = cp.random.random(size = N*N).reshape(N,N)
  np512 = np.random.random(size = N*N).reshape(N,N)
  cp_512.append(cupy_fourier_transform(cp512))
  np_512.append(numpy_fourier_transform(np512))


  N = 1024
  cp1024 = cp.random.random(size = N*N).reshape(N,N)
  np1024 = np.random.random(size = N*N).reshape(N,N)
  cp_1024.append(cupy_fourier_transform(cp1024))
  np_1024.append(numpy_fourier_transform(np1024))


  N = 2048
  cp2048 = cp.random.random(size = N*N).reshape(N,N)
  np2048 = np.random.random(size = N*N).reshape(N,N)
  cp_2048.append(cupy_fourier_transform(cp2048))
  np_2048.append(numpy_fourier_transform(np2048))
```

plotting results

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
fig, ax = plt.subplots()
ax.set_yscale('log')
ax.set_ylabel('Calculation time (s)')
ax.set_xlabel('Matrix dimension')
ax.set_xticks([128,256,512,1024,2048])
ax.scatter([128]*10,cp_128,c='r',marker='*',label = 'CuPy')
ax.scatter([128]*10,np_128,c='b',marker='o',label = 'Numpy')
ax.scatter([256]*10,cp_256,c='r',marker='*',label = 'CuPy')
ax.scatter([256]*10,np_256,c='b',marker='o',label = 'Numpy')
ax.scatter([512]*10,cp_512,c='r',marker='*',label = 'CuPy')
ax.scatter([512]*10,np_512,c='b',marker='o',label = 'Numpy')
ax.scatter([1024]*10,cp_1024,c='r',marker='*',label = 'CuPy')
ax.scatter([1024]*10,np_1024,c='b',marker='o',label = 'Numpy')
ax.scatter([2048]*10,cp_2048,c='r',marker='*',label = 'CuPy')
ax.scatter([2048]*10,np_2048,c='b',marker='o',label = 'Numpy')
by_label = {}

handles, labels = ax.get_legend_handles_labels()
by_label.update(dict(zip(labels, handles)))
ax.legend(by_label.values(), by_label.keys())
plt.show()
```
